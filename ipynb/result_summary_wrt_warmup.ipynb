```python
import torch
import os
import numpy as np
```

# ImageNet30 with 224x224

```python
folder_path = 'result/imgnet30_224_varying_warmup/'
files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
print(files[0])
checkpoint = torch.load(folder_path+files[0])
print(checkpoint.keys())

group_results = {}
groups = ['Uniform', 'kCenterGreedy', 'Uncertainty', 'Forgetting', 'GraNd', 'GradMatch', 'Submodular']
se_dict = {'10':0, '20':1, '40':2, '80':3}
n_se = len(se_dict)
n_exp = 3
n_metric = 2
for g in groups:
    group_results[g] = np.zeros((n_se, n_exp, n_metric))

for i, f in enumerate(files):
    group = f.split('_')[2]
    se = f.split('_')[4].split('se')[1]
    exp = int(f.split('_')[3].split('exp')[1])
    
    print(group, se, exp)

    checkpoint = torch.load(folder_path+f)

    group_results[group][se_dict[se], exp, 0] = checkpoint['best_acc1']
    group_results[group][se_dict[se], exp, 1] = checkpoint['elapsed_time']
```

```python
result = {}
for g in groups:
    for se in se_dict.keys():
        done = (group_results[g][se_dict[se]][:,0].sum() != 0)
        if done:
            name = g+', '+se
            result[name] = group_results[g][se_dict[se]].mean(axis=0)

for r in result:
    print("Method: {}, Acc: {:.2f}, Time: {:.2f}".format( r, result[r][0], result[r][1]))
#print(result)
```
