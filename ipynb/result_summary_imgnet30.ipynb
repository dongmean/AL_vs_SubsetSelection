```python
import torch
import os
import numpy as np
```

# ImageNet30 with 224x224 & ResNet18

```python
folder_path = 'result/imgnet30_224/'
files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
print(files[0])

group_results = {}
groups = ['Uniform', 'kCenterGreedy', 'Uncertainty', 'Forgetting', 'GraNd', 'GradMatch', 'Submodular']
fr_dict = {'0.01':0, '0.05':1, '0.1':2, '0.2':3, '0.4':4}
n_fr = len(fr_dict)
n_exp = 3
n_metric = 2
for g in groups:
    group_results[g] = np.zeros((n_fr, n_exp, n_metric))

for i, f in enumerate(files):
    group = f.split('_')[2]
    fr = f.split('_')[6].split('fr')[1]
    exp = int(f.split('_')[3].split('exp')[1])
    
    print(group, fr, exp)

    checkpoint = torch.load(folder_path+f)

    group_results[group][fr_dict[fr], exp, 0] = checkpoint['best_acc1']
    group_results[group][fr_dict[fr], exp, 1] = checkpoint['elapsed_time']
```

```python
folder_path = 'result/imgnet30_224/'
files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
print(files[0])

for i, f in enumerate(files):
    group = f.split('_')[2]
    fr = f.split('_')[6].split('fr')[1]
    exp = int(f.split('_')[3].split('exp')[1])
    
    if group == 'Forgetting':
        checkpoint = torch.load(folder_path+f)
        forgetting_events = checkpoint['subset']['scores']
        #print(forgetting_events[:100])
        break
```

```python
checkpoint.keys()
```

```python
result = {}
for g in groups:
    for fr in fr_dict.keys():
        done = (group_results[g][fr_dict[fr]][:,0].sum() != 0)
        if done:
            name = g+', '+fr
            result[name] = group_results[g][fr_dict[fr]].mean(axis=0)

for r in result:
    print("Method: {}, Acc: {:.2f}, Time: {:.2f}".format( r, result[r][0], result[r][1]))
#print(result)
```

# ImageNet30 with 32x32 & ResNet9/VGG11/LeNet32

```python
folder_path = 'result/imgnet30_32_arch/'
files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]
print(files[0])
checkpoint = torch.load(folder_path+files[0])
print(checkpoint.keys())

group_results = {}
methods = ['kCenterGreedy', 'Uncertainty', 'Forgetting', 'GraNd', 'GradMatch', 'Submodular']
archs = ['ResNet9', 'VGG11', 'LeNet']
fr_dict = {'0.01':0, '0.05':1, '0.1':2, '0.2':3, '0.4':4, '0.6':5}
n_fr = len(fr_dict)
n_exp = 1
n_metric = 3
for g in methods:
    for a in archs:
        name = g+'_'+a
        group_results[name] = np.zeros((n_fr, n_exp, n_metric))

for i, f in enumerate(files):
    group = f.split('_')[2]
    fr = f.split('_')[7].split('fr')[1]
    exp = int(f.split('_')[3].split('exp')[1])
    arch = f.split('_')[6].split('arch')[1]
    #print(group, fr, exp, arch)

    checkpoint = torch.load(folder_path+f, map_location='cpu')
    #print(checkpoint.keys())
    
    name = group+'_'+arch
    print(name)
    group_results[name][fr_dict[fr], exp, 0] = checkpoint['best_acc1']
    group_results[name][fr_dict[fr], exp, 1] = checkpoint['elapsed_time']
    group_results[name][fr_dict[fr], exp, 2] = checkpoint['warmup_acc']
```

```python
print(group_results['kCenterGreedy_ResNet9'])
```

```python
result = {}
for g in methods:
    for a in archs:
        for fr in fr_dict.keys():
            name = g+'_'+a
            done = (group_results[name][fr_dict[fr]][:,0].sum() != 0)
            if done:
                name2 = name+'_'+fr
                result[name2] = group_results[name][fr_dict[fr]][0]

for r in result:
    print("Method: {}, Acc: {:.2f}, Time: {:.2f}, Warm_acc: {:.2f}".format(r, result[r][0], result[r][1], result[r][2]))
#print(result)
```
